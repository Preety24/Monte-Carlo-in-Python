```python
%matplotlib inline
import import_ipynb
import numpy as np
import random
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set()
N=1000                                         #number of trials

a = np.random.uniform(0,1,N)                   #genrating random generating
f = a**2                                       #defining function for random no a

b = a+1                                        
g = b**2                                       #defining function for random number

plt.scatter(a,f,s=0.5,color='red')             #plotting points separatels 
plt.scatter(b,g,s=0.5,color='blue')
plt.xlabel('x ')
plt.ylabel('f(x)')
plt.show()
```
