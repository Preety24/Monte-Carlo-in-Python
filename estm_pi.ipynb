```python
import import_ipynb
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
N = 20000
x = np.random.uniform(-1,1,N)
y = np.random.uniform(-1,1,N)

inside = x**2 + y**2 <=1

apx_pi = 4*np.sum(inside)/N

print('approx value of pi:',apx_pi)

x_in = x[inside]
y_in = y[inside]

plt.figure(figsize=[5,5])
plt.scatter(x,y,s=1, c='blue')
plt.scatter(x_in,y_in,s=1,c='red')

plt.show()
```
